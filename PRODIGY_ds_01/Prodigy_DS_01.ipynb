```python
#Extracting relevant information
countries = df['Country/Territory']
populations = df['2022 Population']
```

```python
#Sorting the data based on population (descending order)
sorted_data = sorted(zip(countries, populations), key=lambda x: x[1], reverse=True)
#Selecting the 5 most populated countries
top_countries = sorted_data[:5]
top_countries
```

```python
#Plotting a horizontal bar chart for the 5 most populated countries
plt.barh([country[0] for country in top_countries], [population[1] for population in top_countries])
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 5 Countries by Population')
plt.show()
```

```python
#Sorting the data based on population (ascending order)
sorted_data = sorted(zip(countries, populations), key=lambda x: x[1])
#Selecting the 5 least populated countries
least_populated_countries = sorted_data[:5]
least_populated_countries
```

```python
#Plot a horizontal bar chart for the 5 least populated countries
plt.barh([country[0] for country in least_populated_countries], [population[1] for population in least_populated_countries])
plt.xlabel('Population')
plt.title('Least 5 Populated Countries')
plt.show()
```
